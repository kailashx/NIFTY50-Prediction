```python
def divide_dict_seq(dict1):
  """Divide the input dictinary into multiple seq with specified length.

  Args:
    dict1: A dictionary.

  Returns:
    A list with divided seq.
  """
  max_seq_len = 1000
  seq = list()
  tot_seq = list()
  for key, value in dict1.items():
    length = len(key)+len(value)

    if max_seq_len >= length:
      seq.append(str(key)+" is "+str(value))
      max_seq_len -= length
    else
      max_seq_len = 1000
      tot_seq.append(seq)
      seq = list()
      seq.append(str(key)+" is "+str(value))
      max_seq_len -= length
  del seq
  return tot_seq
```

Preparing data for training tokenizer

```python
def remove_all_numeric_data_from_dict(dict1):
  """Removes all numeric data from a dictionary.

  Args:
    dict1: A dictionary.

  Returns:
    A dictionary with all numeric data removed.
  """

  new_dict = {}
  for key, value in dict1.items():
    if not isinstance(value, (int, float)):
      new_dict[key] = value
  return new_dict
```

```python
def dict_to_string(dict1):
  """Converts a dictionary to a string.

  Args:
    dict1: A dictionary.

  Returns:
    A string representation of the dictionary.
  """

  string1 = ""
  for key, value in dict1.items():
    string1 += f"{key}: {value}\n"
  return string1
```

```python
import yfinance as yf
import traceback
def create_text(symbol, f):
  try:
    print(symbol)
    msft = yf.Ticker(symbol)

    # get all stock info
    data = remove_all_numeric_data_from_dict(msft.info)
    #type(msft.info)
    #msft.info.values()
    if 'companyOfficers' in data:
      # inner dict numeric values
      temp = list()
      for officer in data['companyOfficers']:
        officer = remove_all_numeric_data_from_dict(officer)
        temp.append(officer)

        # remove company officers data from old data
      del data['companyOfficers']
      data['companyOfficers'] = temp
      del temp

    # Write the data to the text file
    f.write(dict_to_string(data))
  except Exception as e:
    print('System Exception', e)
    traceback.print_exc()

```

Preparocessing text for vocabulary

```python
import re

def clean_text(text):
  """Cleans text for BERT tokenizer training.

  Args:
    text: A string.

  Returns:
    A cleaned string.
  """

  # Remove punctuation.
  text = re.sub(r'[^\w\s]', '', text)

  # Lowercase the text.
  # text = text.lower()

  return text

def preprocess_corpus(corpus):
  """Preprocesses a corpus for BERT tokenizer training.

  Args:
    corpus: A list of strings.

  Returns:
    A list of cleaned strings.
  """

  cleaned_corpus = []
  for text in corpus:
    cleaned_corpus.append(clean_text(text))

  return cleaned_corpus
```
