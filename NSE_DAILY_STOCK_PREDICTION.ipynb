```python
#@title News Collection and Save
import random
import time
def get_news():
    client = gnewsclient.NewsClient()
    topics = ['Top Stories',
             'World',
             'Nation',
             'Business',
             'Technology',
             'Science',
             'Health']
    world_events  = []
    for loc in client.locations:
      for topic_ in topics:
          client = gnewsclient.NewsClient(language='en', location=loc, topic=topic_, use_opengraph=True, max_results=108)
          news = client.get_news() # list containes dict
          for article in news:
            world_events.append(article['title']+','+ '' if article['description'] is None else article['description'])
            time.sleep(random.uniform(0.0, 0.9)) # wait before getting news again
            print(article['title'])
    return world_events

world_events = get_news()
#print("No of news: ", len(news), ", News: At Index[0] ", news[0], ", At Index[", len(news), "] ", news[len(news)-1])
len(world_events)

```

```python
#@title saving news to file

NEWS_PATH += current_time+'.csv'
with open(NEWS_PATH, 'w') as f:
    for news in world_events:
        f.writelines(news+',')
```

```python
#@title Collect Stock Price and Save
import yfinance as yf

def get_fundamental_info(ticker_name):
  """Divide the input dictinary into multiple seq with specified length.

  Args:
    ticker_name: A SYMBOL.

  Returns:
    info: A dict with basic information about the given symbol.
    hist: historical market data of period 1 day and interval of 1 min
  """
  # get basic data
  ticker = yf.Ticker(ticker_name)

  # get historical market data
  hist = ticker.history(period='1d', interval="1m")
  info = ticker.info
  return hist, info
#hist = get_fundamental_info('WIPRO.NS')

# importing os module
import os
import pickle

# Directory
directory = current_time

# Parent Directory path
parent_dir = STOCK_INFO

# Path
path = os.path.join(parent_dir, directory)

# Create the directory
# 'GeeksForGeeks' in
# '/home / User / Documents'
os.mkdir(path)
print("Directory '% s' created" % directory)
print("Fetching EQUITY Symbols")
for symbol in eq_list:
    hist, info = get_fundamental_info(symbol+'.NS')
    hist.to_csv(path+'/'+symbol+'.csv')
    with open(path+'/'+symbol+'_info.bin', 'wb') as f:
        pickle.dump(info, f)
print("=================================================\nFetching SME Symbols")
for symbol in sme_lis:
    hist, info = get_fundamental_info(symbol+'.NS')
    hist.to_csv(path+'/'+symbol+'.csv')
    with open(path+'/'+symbol+'_info.bin', 'wb') as f:
        pickle.dump(info, f)
```
