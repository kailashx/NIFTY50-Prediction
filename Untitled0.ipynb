```python
class node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

class linked_list:
    def __init__(self):
        self.head = node()

    def append(self, data):
        new_node = node(data)
        cur = self.head
        while cur.next != None:
            cur = cur.next
        cur.next = new_node

    def length(self):
        cur = self.head
        total = 0
        while cur.next != None:
            cur = cur.next
            total += 1
        return total

    def display(self):
        cur = self.head
        data = []
        while cur.next != None:
            data.append(cur.data)
            cur = cur.next
        return data


l_list = linked_list()
```

```python
l_list.append(5)
l_list.append(6)
l_list.append(7)
l_list.append(8)
l_list.append(9)
print(l_list.display())
```
