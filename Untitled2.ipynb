```python
import json
from datetime import date, datetime
c = ticker.calendar
# Define a custom encoder for dates
def json_serial(obj):
    """JSON serializer for objects not serializable by default json code"""
    if isinstance(obj, (datetime, date)):
        return obj.isoformat()
    raise TypeError ("Type %s not serializable" % type(obj))

# Use the custom encoder when dumping to JSON
json.dumps(c, indent=4, default=json_serial)

with open('TICKER.json', 'w') as fp:
    json.dump(ticker.calendar, fp, default=json_serial)
```

```python
from yfinance.exceptions import YFChartError, YFInvalidPeriodError, YFPricesMissingError, YFTzMissingError
import pandas as pd
def find_max_pd(valid_period):
    # 1d,5d,1mo,3mo,6mo,1y,2y,5y,10y,ytd,max
    if "max" in valid_period:
        return "max"
    elif "10y" in valid_period:
        return "10y"
    elif "5y" in valid_period:
        return "5y"
    elif "2y" in valid_period:
        return "2y"
    elif "1y" in valid_period:
        return "1y"
    elif "6mo" in valid_period:
        return "6mo"
    elif "3mo" in valid_period:
        return "3mo"
    elif "1mo" in valid_period:
        return "1mo"
    elif "5d" in valid_period:
        return "5d"
    elif "1d" in valid_period:
        return "1d"
    else:
        return "ytd"
hist = pd.DataFrame()
try:
    hist = ticker.history(period="max")
    metadict = ticker.get_history_metadata()
    if metadict:
        validRanges = metadict['validRanges']
        max_prd = find_max_pd(validRanges)
        # If max is not in ValidRanges
        if max_prd != "max":
            print(f"Retrying with valid period: {validRanges} and max: {max_prd}")
            hist = ticker.history(period=max_prd)

    if not hist.empty:
        from_to = hist.iloc[[0, -1]].index.strftime('%Y_%m_%d').tolist()
        file_name = f'HIST_{from_to[0]}_{from_to[1]}.csv'
        print('DRSCARGO', from_to)
        hist.to_csv(file_name)
except (YFChartError, YFInvalidPeriodError, YFPricesMissingError, YFTzMissingError) as e:
    print(e)
```
